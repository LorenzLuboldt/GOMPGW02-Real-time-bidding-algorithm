# Data Analysis whole dataset

```python
rtb.info()
```

```python
#Number of rows where payprice == 0
print("Number of rows where payprice ==0")
print(len(rtb[rtb['payprice']==0]))
rtb[rtb['payprice']== 0].groupby(['slotid']).size()
```

```python
#Number of Impressions in the data set
print("Number of Impressions in this data set")
Count_Imps = len(rtb)
print(Count_Imps)
```

```python
#Number of clicks and non-clicks in the data set
click = rtb.groupby(['click']).size()
print('Number of clicks (1) and non-clicks (0) in the data set')

print(click)
```

```python
#Number of clicks in the data set
print("Number of Clicks in this data set")
Clicks = rtb.click.nonzero()
Count_Clicks = len(Clicks[0])
print(Count_Clicks)
```

```python
#CTR across the data set
CTR = (Count_Clicks / Count_Imps)
print(CTR)
```

```python
#Total Pay Price across all bids in the data set
Total_Cost = rtb.loc[:,"payprice"].sum()
print("Total Pay Price across all bids in the data set")
print(Total_Cost)
```

```python
#Mean CPM across all impressions in the dataset
Mean_CPM = rtb.loc[:,"payprice"].mean()
print("Mean CPM across all impressions in the dataset")
print(Mean_CPM)
```

```python
#Mean bidprice across all impressions in the dataset
Mean_bidprice = rtb.loc[:,"bidprice"].mean()
print("Mean bid price across all impressions in the dataset")
print(Mean_bidprice)
```

```python
#Highest bidprice across all impressions in the dataset
Max_bidprice = rtb.loc[:,"bidprice"].max()
print("Maximum bid price across all impressions in the dataset")
print(Max_bidprice)
```

```python
#Highest bidprice across all impressions in the dataset
Min_bidprice = rtb.loc[:,"bidprice"].min()
print("Minimum bid price across all impressions in the dataset")
print(Min_bidprice)
```

```python
#Mean slotprice across all impressions in the dataset
Mean_slotprice = rtb.loc[:,"slotprice"].mean()
print("Mean slot price across all impressions in the dataset")
print(Mean_slotprice)
```

```python
#eCPC (mean CPC) across all impressions in the dataset
eCPC = (Mean_CPM/1000)/CTR
print("eCPC across all impressions in the dataset")
print(eCPC)
```

```python
#Show all clicked impressions
clicked_impressions = rtb.loc[rtb['click'] == 1]
clicked_impressions
```

```python
#Show average prices for clicked impressions
average_bidprice = clicked_impressions.loc[:,'bidprice'].mean()
average_payprice = clicked_impressions.loc[:,'payprice'].mean()
print("Average pay prices for clicked impressions")
print(average_payprice)
print("Average bid prices for clicked impressions")
print(average_bidprice)
```

# Data Analysis by Advertisers

```python
# Look at click and non click for each advertiser
df_click = rtb.groupby(['advertiser','click']).size().reset_index(name ='click_or_no_click')
df_click
```

```python
# Calculate number of bids for each advertisers

bid_counts = rtb.groupby(['advertiser']).agg({'bidid':'size'}).rename(columns={'bidid':'number_of_bids'})
bid_counts_sorted = bid_counts.sort_values(by=['number_of_bids'], ascending= False)
bid_counts_sorted
```

```python
advertisers_grouped = rtb.groupby(['advertiser']).mean()
advertisers_grouped
```

```python
#Number of advertisers in the data set
print('Number of advertisers in the data set')
print(len(advertisers_grouped), 'Advertisers')
```

# Data Analysis by Weekday

```python
# Look at click and non click for each weekday
df_click_no_click = rtb.groupby(['weekday','click']).size().reset_index(name ='click_or_no_click').sort_values(by=['click_or_no_click'], ascending= False)
df_click = df_click_no_click[df_click_no_click['click'] == 1].reset_index(drop=True).sort_values(by=['click'], ascending= False)
df_click
```

# Data Analysis by Hour

```python
# Look at click and non click for each day hour
df_click = rtb.groupby(['hour','click']).size().reset_index(name ='click_or_no_click').sort_values(by=['click_or_no_click'], ascending= False)
df_click
```

# Data Analysis by AdExchange

```python
# Look at click and non click for each weekday
df_click = rtb.groupby(['adexchange','click']).size().reset_index(name ='click_or_no_click').sort_values(by=['click_or_no_click'], ascending= False)
df_click
```
